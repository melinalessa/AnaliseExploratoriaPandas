Analise exploratoria do banco "Iris" (planta)

Importar bilbioteca

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Armazenar em uma variavel e ler o arquivo

```python
df= pd.read_csv('Iris.csv', sep= ",")
df
```

Excluir a coluna ID

```python
df.drop(columns= 'Id', inplace= True)
df
```

Analisar os tipos de especies

```python
df['Species'].unique()
```

Criamos uma variavel para especies e analisar

```python
frequency_iris= df['Species'].value_counts()
frequency_iris
```

Criando grafico somente das espécies na vertical

```python
plt.bar(x=frequency_iris.index, height=frequency_iris)
plt.title('Frequencia das Espécies das Iris')
plt.ylim(0,80)
plt.show()
```

Criando grafico somente das espécies na horizontal

```python
plt.barh(y=frequency_iris.index, width=frequency_iris)
plt.title('Frequencia das Espécies das Iris')
plt.xlim(0,80)
plt.show()
```

Criando grafico de Pizza

```python
explode= (0,0.1,0.1)
labels= frequency_iris.index
plt.pie(frequency_iris, explode=explode, shadow=True, autopct="%1.2f%%", labels=labels)
plt.title('Pizza Chart', fontsize=15)
plt.legend(labels, loc= 'upper left')
plt.show()
```
